### Load dataset

```python
#First, load the required libraries:

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Now, load the dataset with pandas:
df = pd.read_csv('D:\My personal projects\DATASETS\Crime_Data_from_2020_to_Present.csv')

#Finally, view the data:
df.head()
```

### Basic EDA

```python
#Show all columns:
pd.set_option('display.max_columns', None)
df.head()
```

```python
#To understand our data, we are going to rename the necessary columns:
df=df.rename(columns={'DR_NO':'Number district'})
df=df.rename(columns={'Date Rptd':'Crime reported date'})
df=df.rename(columns={'DATE OCC':'Crime occurrence date'})
df=df.rename(columns={'TIME OCC':'Time occurrence'})
df=df.rename(columns={'Rpt Dist No':'Reported discrict number'})
df=df.rename(columns={'Crm Cd':'Crime code'})
df=df.rename(columns={'Crm Cd Desc':'Crime Description'})
df=df.rename(columns={'Mocodes':'Modus Operandi'})
df=df.rename(columns={'Weapon Used Cd':'Weapon Used Code'})
df=df.rename(columns={'Crm Cd 1':'Crime code 1'})
df=df.rename(columns={'Crm Cd 2':'Crime code 2'})
df=df.rename(columns={'Crm Cd 3':'Crime code 3'})
df=df.rename(columns={'Crm Cd 4':'Crime code 4'})
```

### Quick plots

```python
df['Vict Sex'].value_counts()
```

```python
#Crimes by sex in pie graph:
df['Vict Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Crimes by sex')
plt.show()
```

```python
#Crimes by sex in bar graph:
plt.figure(figsize=(10,10))
sns.countplot(x='Vict Sex', data=df)
plt.title('Crimes by sex')
plt.show()
```

```python
df['Vict Descent'].value_counts()
```

```python
#Crimes by descent in pie graph:
df['Vict Descent'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Crimes by descent')
plt.show()
```

```python
#Crimes by descent in bar graph:
plt.figure(figsize=(10,10))
sns.countplot(x='Vict Descent', data=df)
plt.title('Number of crimes per descent')
plt.show()
```
